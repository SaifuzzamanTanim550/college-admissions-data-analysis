```python
# Calculate admit/deny counts per school
school_decisions = (
    df.groupby(["hs_last_attended_name", "recommendation"])
      .size()
      .unstack(fill_value=0)
)

for col in ["ADMIT", "DENY"]:
    if col not in school_decisions.columns:
        school_decisions[col] = 0

# Calculate total applications and acceptance rate
school_decisions["total_apps"] = school_decisions["ADMIT"] + school_decisions["DENY"]
school_decisions["accept_rate"] = school_decisions["ADMIT"] / school_decisions["total_apps"]

# Filter schools with minimum applications
min_apps = 30
filtered = school_decisions[school_decisions["total_apps"] >= min_apps].copy()

# Separate schools by acceptance rate (perfect vs. non-perfect)
perfect = filtered[filtered["accept_rate"] == 1].copy()
nonperfect = filtered[filtered["accept_rate"] < 1].copy()

# Sort perfect acceptance schools by admit count
perfect = perfect.sort_values("ADMIT", ascending=False)

# Sort non-perfect acceptance schools by acceptance rate
nonperfect = nonperfect.sort_values("accept_rate", ascending=False)

# Combine and get top 10 schools
top_schools = pd.concat([perfect, nonperfect]).head(10).reset_index()
top_schools.rename(columns={"hs_last_attended_name": "high_school"}, inplace=True)

# Define function to shorten school names
def shorten_name(name, max_len=35):
    return name if len(name) <= max_len else name[:max_len-3] + "..."

top_schools["high_school_short"] = top_schools["high_school"].apply(shorten_name)

# Create labels for bars
top_schools["label"] = top_schools.apply(
    lambda row: f"{row['accept_rate']:.1%} ({row['ADMIT']}/{int(row['total_apps'])})",
    axis=1
)

professional_colors = [
    "#2A4B7C",
    "#4F6D7A",
    "#6C757D",
    "#5A8F7B",
    "#8C7A9C",
    "#A67F58",
    "#7F8C8D",
    "#3C6E71",
    "#9A9EAB",
    "#556B8E"
]

# Plot the bar chart
fig = px.bar(
    top_schools,
    x="high_school_short",
    y="accept_rate",
    text="label",
    title="Top High Schools by Acceptance Quality",
    color="high_school_short",
    color_discrete_sequence=professional_colors,
    hover_data={
        "high_school": True,
        "accept_rate": ":.2%",
        "ADMIT": True,
        "total_apps": True,
        "high_school_short": False
    }
)

fig.update_traces(textposition="outside")

fig.update_layout(
    template="plotly_white",
    xaxis_title="High School",
    yaxis_title="Acceptance Rate",
    yaxis_tickformat=".0%",
    font=dict(size=13),
    bargap=0.35,
    showlegend=False,
    xaxis_tickangle=45,
    margin=dict(b=120)
)

fig.show()
```

```python
# Create binary admit column
df["admit_binary"] = (df["recommendation"] == "ADMIT").astype(int)

# Define GPA bins and labels
bins = [0, 70, 80, 85, 90, 95, 110]
labels = ["Below 70", "70–80", "80–85", "85–90", "90–95", "95–100"]

df["gpa_band"] = pd.cut(
    df["overall_gpa"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Calculate admit count, total students, and acceptance rate per GPA band
band_rates = (
    df.groupby("gpa_band", observed=True)
      .agg(
          admit_count=("admit_binary", "sum"),
          total_students=("admit_binary", "count")
      )
      .reset_index()
)

band_rates["acceptance_rate"] = (
    band_rates["admit_count"] / band_rates["total_students"]
)

# Create labels for bars
band_rates["rate_label"] = band_rates["acceptance_rate"].map(lambda x: f"{x:.1%}")

band_color_map = {
    "Below 70": "#b71c1c",
    "70–80":    "#e65100",
    "80–85":    "#f9a825",
    "85–90":    "#7cb342",
    "90–95":    "#2e7d32",
    "95–100":   "#1b5e20"
}
bar_colors = band_rates["gpa_band"].astype(str).map(band_color_map).tolist()

# Create bar chart
fig = px.bar(
    band_rates,
    x="gpa_band",
    y="acceptance_rate",
    text="rate_label",
    title="Freshman Admission Rates by High School GPA Range",
    labels={
        "gpa_band": "High School GPA Range",
        "acceptance_rate": "Chance of Admission"
    },
    category_orders={"gpa_band": labels}
)

fig.update_traces(
    marker_color=bar_colors,
    textposition="outside",
    customdata=band_rates[["admit_count", "total_students"]].to_numpy(),
    hovertemplate=(
        "<b>%{x}</b><br>" +
        "Chance of admission: %{y:.1%}<br>" +
        "Admitted: %{customdata[0]} of %{customdata[1]} students" +
        "<extra></extra>"
    )
)

# Update layout
fig.update_layout(
    template="plotly_white",
    yaxis_tickformat=".0%",
    xaxis_title="High School GPA Range",
    yaxis_title="Chance of Admission",
    font=dict(size=15),
    bargap=0.35,
    showlegend=False
)

fig.show()
```

```python
# Create binary admit target
df["admit_binary"] = df["recommendation"].apply(lambda x: 1 if x == "ADMIT" else 0)

def plot_logistic_for_grade(df, grade_col, target_col="admit_binary"):
    """
    Fit and plot a logistic regression model for one grade column.
    grade_col: name of the GPA column (e.g., 'overall_gpa', 'math_gpa')
    """
    # Drop rows with missing values
    data = df.dropna(subset=[grade_col, target_col]).copy()

    X = data[[grade_col]].values
    y = data[target_col].values

    # Scale feature
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Fit logistic regression model
    model = LogisticRegression()
    model.fit(X_scaled, y)

    # Generate GPA range for prediction curve
    gpa_min, gpa_max = data[grade_col].min(), data[grade_col].max()
    gpa_range = np.linspace(gpa_min, gpa_max, 300).reshape(-1, 1)
    gpa_range_scaled = scaler.transform(gpa_range)
    probs = model.predict_proba(gpa_range_scaled)[:, 1]

    # Create DataFrame for the curve
    curve_df = pd.DataFrame({
        grade_col: gpa_range.flatten(),
        "prob_admit": probs
    })

    # Calculate GPA cutoff for 50% admission probability
    z_cutoff = -model.intercept_[0] / model.coef_[0][0]
    gpa_cutoff = scaler.inverse_transform([[z_cutoff]])[0][0]

    # Plot admission probability curve
    fig = px.line(
        curve_df,
        x=grade_col,
        y="prob_admit",
        title=f"Logistic Regression: Probability of Admission vs {grade_col}",
        labels={grade_col: grade_col.replace("_", " ").title(),
                "prob_admit": "Admission Probability"},
        color_discrete_sequence=["#2A4B7C"]
    )

    # Add raw student outcome data points
    fig.add_scatter(
        x=data[grade_col],
        y=data[target_col],
        mode="markers",
        opacity=0.25,
        marker=dict(color="#6C757D"),
        name="Student outcomes (0 = Deny, 1 = Admit)"
    )

    # Add 50% probability cutoff line
    fig.add_vline(
        x=gpa_cutoff,
        line_dash="dash",
        line_color="#B56576",
        annotation_text=f"50% cutoff ≈ {gpa_cutoff:.1f}",
        annotation_position="top left"
    )

    fig.update_layout(
        template="plotly_white",
        yaxis_range=[-0.05, 1.05],
        yaxis_tickformat=".0%",
        font=dict(size=14)
    )

    fig.show()
    print(f"{grade_col}: estimated GPA for 50% admit probability ≈ {gpa_cutoff:.2f}")

# Run logistic regression for each GPA column
grade_cols = ["overall_gpa", "english_gpa", "math_gpa", "science_gpa", "social_studies_gpa"]

for col in grade_cols:
    plot_logistic_for_grade(df, col)
```

```python
# Count applicants per ethnicity
eth_total = (
    df.groupby("ethnicity")
      .size()
      .reset_index(name="applicants")
)

# Convert ethnicity to string type
eth_total["ethnicity"] = eth_total["ethnicity"].astype(str)

# Count admitted students per ethnicity
eth_admit = (
    df[df["recommendation"] == "ADMIT"]
      .groupby("ethnicity")
      .size()
      .reset_index(name="admitted")
)

# Convert ethnicity to string type
eth_admit["ethnicity"] = eth_admit["ethnicity"].astype(str)

# Merge data and calculate denial/acceptance rates
eth = pd.merge(eth_total, eth_admit, on="ethnicity", how="left").fillna(0)
eth["admitted"] = eth["admitted"].astype(int)
eth["denied"] = eth["applicants"] - eth["admitted"]
eth["accept_rate"] = eth["admitted"] / eth["applicants"]

# Calculate percentage of all applicants per ethnicity
total_apps = eth["applicants"].sum()
eth["pct_all"] = eth["applicants"] / total_apps

# Calculate within-group admit/deny percentages
eth["pct_admit_within"] = eth["admitted"] / eth["applicants"]
eth["pct_deny_within"]  = eth["denied"] / eth["applicants"]

# Calculate scaled heights for stacked bars
eth["height_admit"] = eth["pct_all"] * eth["pct_admit_within"]
eth["height_deny"]  = eth["pct_all"] * eth["pct_deny_within"]

# Prepare data for stacked bar chart
long_df = eth.melt(
    id_vars=["ethnicity", "pct_all", "accept_rate", "admitted", "denied", "applicants"],
    value_vars=["height_admit", "height_deny"],
    var_name="decision",
    value_name="height"
)

decision_map = {
    "height_admit": "Admitted",
    "height_deny":  "Denied"
}
long_df["decision"] = long_df["decision"].map(decision_map)

# Define color map
color_map = {
    "Admitted": "#1f4e79",
    "Denied":   "#7f7f7f"
}

# Create stacked bar chart
fig = px.bar(
    long_df,
    x="ethnicity",
    y="height",
    color="decision",
    barmode="stack",
    title="Ethnicity Representation and Admission Split",
    color_discrete_map=color_map,
)

# Update layout
fig.update_layout(
    template="plotly_white",
    yaxis_title="Share of Total Applicants",
    xaxis_title="Ethnicity",
    yaxis_tickformat=".0%",
    font=dict(size=14),
    xaxis_tickangle=35,
    bargap=0.25,
    margin=dict(b=130),
    legend_title_text=""
)

# Update traces with custom hover info
fig.update_traces(
    customdata=long_df[["applicants", "admitted", "denied", "accept_rate"]].to_numpy(),
    hovertemplate=(
        "<b>%{x}</b><br>" +
        "Decision: %{legendgroup}<br>" +
        "Share of total applicants: %{y:.1%}<br>" +
        "<br>" +
        "Total applicants: %{customdata[0]}<br>" +
        "Admitted: %{customdata[1]}<br>" +
        "Denied: %{customdata[2]}<br>" +
        "Acceptance rate: %{customdata[3]:.1%}" +
        "<extra></extra>"
    )
)

fig.show()
```
